### Data Preparation

#### 1. **Data Retrieval**
The data used in this analysis is based on gold trade information sourced from Yahoo Finance, retrieved at 30-minute intervals. The dataset contains the following key variables:
- **Date**: The timestamp for each trading period (30 minutes).
- **Open**: The price of gold at the start of the 30-minute period.
- **Close**: The price of gold at the end of the 30-minute period.
- **High**: The highest price of gold within the 30-minute period.
- **Low**: The lowest price of gold within the 30-minute period.

These variables serve as the fundamental time-series data that is enriched with technical indicators to improve predictive modeling.

#### 2. **Calculation of Technical Indicators**
To better capture the underlying trends and momentum of gold prices, several technical indicators are calculated. These indicators are widely used in financial markets to forecast price movements. The technical indicators calculated include:
- **Moving Averages (e.g., SMA, EMA)**: To track the average price over specific time windows.
- **Relative Strength Index (RSI)**: To measure the magnitude of recent price changes to evaluate overbought or oversold conditions.
- **Bollinger Bands**: To measure volatility and identify potential buy/sell signals based on price movement relative to the bands.
- **Moving Average Convergence Divergence (MACD)**: To track the relationship between two moving averages of a securityâ€™s price.

The dataset containing these technical indicators is stored in the file `raw-data.csv`, where each technical indicator corresponds to the period it describes.

#### 3. **Shifting Technical Indicators to the Next Period**
The key insight in this step is to move the calculated technical indicators from the current period to the row representing the next 30-minute period. This transformation ensures that the technical indicators for one period are used to predict the opening, closing, high, and low prices of the next period.

This shifting of indicators aligns the features with the prediction task, as it allows the machine learning model to use the technical indicators from the previous period to predict the target prices in the next period. This process was automated using a custom Python script, `Get raw data.py`.

#### 4. **Creation of the Target Variable: Action**
A new indicator, called **Action**, is introduced in `raw-data.csv` as the target variable (Y). The **Action** variable is derived by calculating the price change between periods. The price change is then categorized into a range from `-6` to `6`, with negative values indicating a decline in price and positive values indicating a rise.

This target classification allows the model to categorize price movements over multiple classes, enabling it to forecast not just whether the price will increase or decrease, but also the magnitude of the change.

- A **positive Action value** (1 to 6) represents varying degrees of price increases.
- A **negative Action value** (-1 to -6) represents varying degrees of price decreases.
- **Zero** represents no significant change in the price.

#### 5. **Final Dataset**
The final dataset consists of:
- The **open, close, high, and low prices** for the upcoming period (next 30-minute interval).
- The **technical indicators** calculated from the previous period.
- The **Action** variable, representing the price movement range as the target column for model training.

This data preparation framework ensures that the dataset is well-structured for supervised learning, where past technical indicators are used to predict future price movements.

### Feature Engineering: Phase 2 - Applying Polynomial Features and Dimensionality Reduction

In the second phase of the project, I expanded the feature space by applying polynomial features of degree 3 to capture more complex interactions between variables. This significantly increased the number of features, from the original set to **26,000 features**, which posed computational challenges due to the high dimensionality of the data.

#### Challenge:
The sheer number of features (26,000) made it impractical to compute and train models efficiently. Processing such a large feature space would require extensive computational resources and time, making it necessary to apply dimensionality reduction techniques.

#### Solution: PCA Analysis
To overcome this issue, I utilized **Principal Component Analysis (PCA)**, which is a dimensionality reduction technique that transforms the data into a set of orthogonal components while retaining as much variance as possible. Specifically, I set the PCA to cover **99% of the variance** in the data, ensuring that we maintained almost all of the information from the original feature space.

- **Original Features**: 26,000 features after polynomial expansion.
- **Post-PCA Features**: 79 principal components were sufficient to describe 99% of the variance in the dataset.

By applying PCA, the dimensionality of the data was reduced to **79 components**, making it computationally feasible to continue the model training process. This allowed the models to focus on the most important features while maintaining high information coverage, ultimately improving efficiency without sacrificing model performance.

```python
# GRIDSEARCH
import time
times=[]
for model_name, (model, params) in models.items():
    start_time = time.time()
    grid = GridSearchCV(model, param_grid=params, scoring='f1', cv=5, n_jobs=-1)  # Use F1-score for imbalanced data
    grid.fit(X_train_res, y_train_res)  # Train with oversampled data (SMOTE)
    best_model = grid.best_estimator_
    best_models[model_name] = best_model
    end_time = time.time()
    execution_time = end_time - start_time
    times.append(f"{model_name} execution time: {execution_time} seconds")
    print(f"{model_name} execution time: {execution_time} seconds")
    # Save the best model using joblib
    joblib.dump(best_model, f'{model_name}_best_model.pkl')

    print(f"{model_name} best parameters: {grid.best_params_}")
```

```python
# Evaluate the best models on the test set
for model_name, model in best_models.items():
    y_pred = model.predict(X_test)
    print(f"\n{model_name} Results:")
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(f"F1-Score: {f1_score(y_test, y_pred):.4f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

# Plot comparison of model performance using F1-score
f1_scores = [f1_score(y_test, model.predict(X_test)) for model in best_models.values()]
model_names = list(best_models.keys())
plt.figure(figsize=(10, 6))
plt.barh(model_names, f1_scores, color='skyblue')
plt.xlabel('F1-Score')
plt.title('Model Comparison using F1-Score')
plt.show()

print('ENDING NOW')
```
